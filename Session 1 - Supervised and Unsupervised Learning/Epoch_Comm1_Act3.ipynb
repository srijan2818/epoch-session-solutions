## Gaussian Mixture Models
Let's now look at how we can go about implementing one of the topics discussed: Gaussian Mixture Models

```python
# Import the relevant libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

np.random.seed(42)
```

```python
# Creating the dataset for us to show the clustering on

X, y_true = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=[1.0, 2.5, 0.8],
    random_state=42
)

print("Dataset shape:", X.shape)
```

```python
# As mentioned before, visualising the dataset is almost always the right thing to do in the beginning

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], s=15)
plt.title("Unlabeled Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Performing Gaussian Mixture operation
# Go on and change parameters as you see fit, get a good sense of how its doing what its doing

gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm.fit(X)

labels = gmm.predict(X)
means = gmm.means_

print("Learned means:\n", means)
```

```python
# Lets see what we have now

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=15, cmap='viridis')
plt.scatter(means[:, 0], means[:, 1], c='red', s=100, marker='x')
plt.title("GMM Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
